```python
import numpy as np

file_path = "/content/drive/My Drive/students.csv"

data = np.genfromtxt(
    file_path,
    delimiter=',',
    skip_header=1
)

print(data)
```

```python
clean_data = data[~np.isnan(data)]
clean_data
```

```python
ids=data[:,0]
marks=data[:,1]
age=data[:,2]
```

```python
#mean_value = np.nanmean(data)
#data[np.isnan(data)] = mean_value
#data
mean_value = np.nanmean(marks)
marks[np.isnan(marks)] = mean_value
data
```

```python
float(np.nanmean(marks))
```

```python

float(np.nanmax(marks))
```

```python
high_scorers = ids[marks > 80]
print("High scorers:", high_scorers)
```

```python
avg_scorers = ids[marks > 50]
print("Average scorers:", avg_scorers)
```

```python
young_students = ids[age < 21]
print("Young students:", young_students)
```

```python
print("25th percentile:", np.percentile(marks, 25))
print("Median (50th percentile):", np.percentile(marks, 50))
print("75th percentile:", np.percentile(marks, 75))
```

```python
sorted_marks = np.sort(marks)
print("Sorted Marks: ", sorted_marks)
```

```python
import matplotlib.pyplot as plt

plt.hist(marks, bins=5, color='skyblue', edgecolor='black')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.show()
```
