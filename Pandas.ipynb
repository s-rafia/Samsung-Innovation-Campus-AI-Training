# **DataFrame**

**Creating DataFrames - Dictioneries**

```python
data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Age': [20, 22, 21],
    'Marks': [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

**From List of Lists**

```python
data = [
    ['Ali', 20, 85],
    ['Sara', 22, 90],
    ['Ahmed', 21, 88]
]

df = pd.DataFrame(data, columns = ['Name', 'Age', 'Marks'])
print(df)
```

# **Handling Missing Data**

**Create Missing Data**

```python
import numpy as np

data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Marks': [85, np.nan, 90]
}

df = pd.DataFrame(data)
df
```

**Detect Missing Values**

```python
df.isnull()
```

```python
df.isnull().sum()
```

**Remove Missing Data**

```python
df.dropna()
```

Fill Missing Data

```python
df.fillna(0)
```

```python
df['Marks'].fillna(df['Marks'].mean())
```
