Polymorphism

```python
# Parent Class
class Shape:
  def area(self):
    print("This is a shape")

#child class 1
class Circles(Shape):
  def area(self):
    print("Area of a Circle")

#child class 2
class Square(Shape):
  def area(self):
    print("Area pf a Square")

# Create Objects
C = Circles()
S = Square()

#Call method directly

C.area()
S.area()
```

```python
# Creating Class
class Multiplication:
  def mul(self, x, y=1):
    return x*y

# Method overloading
Multiply = Multiplication()

print(Multiply.mul(4))
print(Multiply.mul(4, 5))
```

```python
class Animal:
  def speak(self):
    print("Animal makes sound")

class Dog(Animal):
  def bark(self):
    print("Dog barks")

class Cat(Animal):
  def speak(self):
    print("Cat Meow")

d = Dog()
d.speak()
d.bark()

c = Cat()
c.speak()
```

```python
from abc import abstractmethod
class Shape:
  @abstractmethod
  def area(self):
    pass

# Child Class 1
class Rectangle(Shape):

  def __init__(self, width, length):
    self.width = width
    self.length = length

  def area(self):
    return self.width * self.length

# Create Objects
R = Rectangle(5,6)

# Call method directly
print("Rectangle Ares: ", R.area())
```
