# **Encapsulation**

```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance

acc = BankAccount(100)
acc.deposit(50)

acc.get_balance()
```

# **Inheritance**

1. Single Inheritance

```python
class Animal:
  def speak(self):
    print("Animal makes sound")

class Dog(Animal):
  def bark(self):
    print("Dog barks")

d = Dog()
d.speak()
d.bark()
```

2. Multilevel Inheritance

```python
class Animal:
  def eat(self):
    print("Eating")

class Mammal(Animal):
  def walk(self):
    print("Walking")

class Dog(Mammal):
  def bark(self):
    print("Barking")

d = Dog()
d.eat()
d.walk()
d.bark()
```

3. Hierarchical Inheritance

```python
class Animal:
  def speak(self):
    print("Animal speaks")

class Dog(Animal):
  pass

class Cat(Animal):
  pass
```

4. Multiple Inheritance

```python
class Father:
  def skills(self):
    print("Gardening")

class Mother:
  def talents(self):
    print("Cooking")

class Child(Father, Mother):
  pass

c = Child()
c.skills()
c.talents()
```

5. Hybrid Inheritance

```python
class Vehicle: # Grandparent
  def start(self):
    print("Vehicle started")

class Car(Vehicle): # Multilevel part
  def drive(self):
    print("Car is driving")

class ElectricDevice: # Another Parent
  def charge(self):
    print("Charging battery...")

# Hybrid: Inherits from car (Which inherits from Vehicle) And ElectricDevice
class Tesla(Car, ElectricDevice):
  pass

my_tesla = Tesla()
my_tesla.start() # From Vehicle
my_tesla.drive() # From Car
my_tesla.charge() # From ElectricDevice
```

```python
class Person:
  def human(self):
    print("This is a human")

class Employee(Person):
  def job(self):
    print("This person is a company's employee")

class Manager(Employee):
  def work(self):
    print("This person works as a Manager at a company")

m = Manager()
m.human()
m.job()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

class Student(Person):
  def __init__(self, name, age, grade):
    Person.__init__(self, name, age)
    self.grade = grade

  def display(self):
    print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

s1 = Student("Ali", 25, "A")
s1.display()
```
