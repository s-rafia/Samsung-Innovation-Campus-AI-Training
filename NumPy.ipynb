# **Linear Algebara - Dot Product**

```python
a = np.array([[1, 2],
             [3, 4]])

b = np.array([[5,6],
              [7,8]])

print(np.dot(a,b))
```

```python
c = np.array([[2, 0],
              [1, 3]])

d = np.array([[1, 2],
              [3, 4]])
print(np.dot(c,d))
```

# **Iteration**

```python
a = np.arange(6).reshape(2,3)

for i in a:
  print(i)
```

nditer - Prints each element individually

```python
for i in np.nditer(a):
  print(i)
```

# **Stacking**

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6],[7,8]])

print(np.hstack((a,b)))
```

```python
print(np.vstack((a,b)))
```
