```python
data = {
    "Product": ["Pen", "Book", "Bag", "Shoes"],
    "Price": [10, 50, 500, 2000],
    "Stock": [100, 40, 20, 10]
}

df = pd.DataFrame(data)
print(df)
```

Products with price greater than 100

```python
price_gt_100 = df[df["Price"] > 100]
print("Products with price > 100: ")
print(price_gt_100)
```

Sort by stock

```python
sort_by_stock = df.sort_values(by="Stock")
print("\nSorted bt Stock:")
print(sort_by_stock)
```

sort by product

```python
sort_by_product = df.sort_values(by="Product")
print("\nSorted bt Product:")
print(sort_by_product)
```

Adding a new column

```python
df["TotalValue"] = df["Price"] * df["Stock"]
print("\nDataFrame with TotalValue Column:")
print(df)
```

Stock less than 30

```python
stock_lt_30 = df[df["Stock"] < 30]
print("\nProducts with Stock < 30")
print(stock_lt_30)
```
