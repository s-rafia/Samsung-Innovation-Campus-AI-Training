```python
class Student:
  def __init__(self, name, marks):
    self.name = name
    self.marks = marks
  def display(self):
    print(f"Name: {self.name}, Marks: {self.marks}")

s1 = Student("Ali", 90)
s2 = Student("Sara", 85)
s1.display()
s2.display()
```

```python
class Car:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year
  def display(self):
    print(f"Brand: {self.brand}, Model: {self.model}, Year: {self.year}")

c1 = Car("Toyota", "Cultus", 2024)
c2 = Car("Honda", "City", 2022)
c1.display()
c2.display()
```

```python
class Menu:
  def __init__(self, item, price, serving):
    self.item = item
    self.price = price
    self.serving = serving
  def display(self):
    print(f"Item: {self.item}, Price: Rs. {self.price}, Serving: {self.serving} people")

m1 = Menu("Chocholate Cake", 1200, 4)
m2 = Menu("Brownie", 600, 1)
m1.display()
m2.display()
```
