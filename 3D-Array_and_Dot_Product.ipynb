# Linear Algebra - Dot Product

Metrix Multiplication: Row x Column rule

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(np.dot(a,b))
```

```python
c = np.array([[2,0],[1,3]])
d = np.array([[1,2],[3,4]])
print(np.dot(c,d))
```

```python
c1 = np.arange(8).reshape(2,2,2)
c1
```

```python
d1 = np.arange(8).reshape(2,2,2)
d1
```

```python
print(np.dot(c1,d1))
```

# Iteration

```python
a = np.arange(6).reshape(2,3)

for i in a:
  print(i)
```

nditer:
Prints each element individually

```python
for i in np.nditer(a):
  print(i)
```

# Stacking

hstack - add columns

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(np.hstack((a,b)))
```

vstack - add rows

```python
print(np.vstack((a,b)))
```
