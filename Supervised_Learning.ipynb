```python
import pandas as pd
```

# 1.1 Load Dataset

```python
data = pd.read_csv("/content/data_galton.csv")
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
%matplotlib inline
```

```python
# Display the dictionary keys.
data.keys()
```

```python
# Display the description on the data.
print(data['child'])
```

```python
# The explanatory variables.
X = data['child']
Y = data['parent']
```

# 1.2 Convert the data into a DataFrame and then explore:

```python
df = pd.DataFrame(data)
df["Sum"] = df['child']+df['parent']
```

```python
df.head()
```

```python
df.tail()
```

```python
# Descriptive statistics of the variables (columns).
df.describe()
```

```python
# Pair-wise correlation matrix.
np.round(df.corr(), 2)
```

```python
# Visualize the correlation matrix.
sns.heatmap(df.corr(), cmap='coolwarm')
plt.show()
```

```python
# Visualize Child vs Parent.
plt.scatter(X,Y,c = 'b',s=15,alpha=0.5)
plt.xlabel("Child")
plt.ylabel("Parent")
plt.show()
```

# 1.3 Train by linear regression:

```python
# Train
lm = LinearRegression(fit_intercept=True)
lm.fit(X.values.reshape(-1,1), Y)
```

```python
# The intercept.
lm.intercept_
```

```python
# The rest of coefficients (parameters).
lm.coef_
```
