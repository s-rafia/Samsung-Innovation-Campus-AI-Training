Series (1D-Index)

```python
import pandas as pd

data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

```python
s = pd.Series([100, 200, 300], index = ['a', 'b', 'c'])
print(s)
```

```python
print(s['a'])
print(s[0])
```

DataFrames-Dictioneries

```python
data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Age': [20, 22, 21],
    'Marks': [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

From list to lists

```python
data = [
    ['Ali', 20, 85],
    ['Sara', 22, 90],
    ['Ahmed', 21, 88]
]
# gave data as lists instead of dictioneries
df = pd.DataFrame(data, columns=['Name', 'Age', 'Marks']) # instead of using word index, we use columns here
print(df)
```

Handling missing data

Create missing data:

```python
import numpy as np

data = {
    'Name': ['Ali', 'Sara', 'Ahmed'],
    'Marks': [85, np.nan, 90]
}

df = pd.DataFrame(data)
df
```

Detecting Missing values

```python
df.isnull()
```

```python
df.isnull().sum()
```

Remove missing data

```python
df.dropna()
```

Fill missing data

```python
df.fillna(0)
```

```python
df['Marks'].fillna(df['Marks'].mean())
```
